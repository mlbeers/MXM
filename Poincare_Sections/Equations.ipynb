```python
from Library import *
import dill
import os
from collections import defaultdict
import glob
```

```python
final_eqs = defaultdict(list)
final_eq_map = {}
i = 0

results_dir = 'results'
for subdir in os.listdir(results_dir):
    full_subdir_path = os.path.join(results_dir, subdir)
    if os.path.isdir(full_subdir_path):
        target_file = os.path.join(full_subdir_path, 'final_eq.dill')
        if os.path.exists(target_file):
            with open(target_file, 'rb') as f:
                eq = dill.load(f)
                if eq not in final_eq_map:
                    final_eq_map[eq] = i
                    i += 1
                final_eqs[final_eq_map[eq]].append(subdir)
final_eqs = dict(final_eqs)
```

```python
cusp_eqs = defaultdict(list)
cusp_eq_map = {}
i = 0

results_dir = 'results'
for subdir in os.listdir(results_dir):
    full_subdir_path = os.path.join(results_dir, subdir)
    if os.path.isdir(full_subdir_path):
        for subdir2 in os.listdir(full_subdir_path):
            full_subdir_path2 = os.path.join(full_subdir_path, subdir2)
            if os.path.isdir(full_subdir_path2):
                dill_files = glob.glob(os.path.join(full_subdir_path2, '*.dill'))
                for target_file in dill_files:
                    if os.path.exists(target_file):
                        with open(target_file, 'rb') as f:
                            eq = dill.load(f)
                            if eq not in cusp_eq_map:
                                cusp_eq_map[eq] = i
                                i += 1
                            cusp_eqs[cusp_eq_map[eq]].append(f"{subdir}, {subdir2}, {target_file.split('/')[-1][:-5]}")
cusp_eqs = dict(cusp_eqs)
```
